## Study of housing prices in Belgium, part 1: webscraping the data

This study aims at predicting the price of houses and appartments (properties) sold in Belgium. For this study, we extract the data from all properties to be sold overall in Belgium on the website immoweb.be. We extract as many characteristics as possible such that we can study the price variations, all over the country.

The data extraction is processed in two parts:
1. First we extract a list of the properties hyperlinks of all the houses and appartements sold on the website.
2. We loop through the list of hyperlinks, and for each of the properties, exrtact all of their parameters (location, price, size, nb bedrooms etc).

```python
# create an object 'url_list' which will contain the list of hyperlinks scraped from the results web page.
url_list = []
# create an object 'property_id' which will contain the first characteristics of each property
property_id = []

for count, code in enumerate(zipcode):
  page = 1
  data = [0]
  print(count,"/", len(zipcode)-1)
  while data != []:
    try:
        # extract the html code for the page of the search for the zipcode
        url ="https://www.immoweb.be/fr/recherche/maison-et-appartement/a-vendre/{}/{}?countries=BE&page={}".format("*",code,page)
        soup = BeautifulSoup(requests.get(url).content, "html.parser")
        # create an object 'data' which reads json format and saves all data that corresponds to the "results" of the search.
        data = json.loads(soup.find("iw-search")[":results"])
        # for each page of data collected, we extract the hyperlink to each property and save it in a list.
        # we also extract some first hand data for each property and save it in property_id
        for ad in data:
            hyperlink = "https://www.immoweb.be/fr/annonce/{}".format(ad["id"])
            url_list.append(hyperlink)
            property_id.append({"id":ad["id"],"Type":ad["property"]["type"],"Prix":ad["transaction"]["sale"]["price"], "Code postal": ad["property"]["location"]["postalCode"]})
        page += 1
    except:
        print(data)
```

- The list of property id's contains the primary information about each property scraped from the results webpage (ID, property type, price and zip code).
- The list of urls contains only the urls to information about each property sold on the website

```python
# save the lists to excel files

def save_url(list):
    # transform the list into a pandas Series
    url_list_se = pd.Series(list)
    # drop duplicated observations
    url_list_se.drop_duplicates(inplace = True)
    # save data to excel
    url_list_se.to_excel('url_list_' + date_string + '.xls')

save_url(url_list)

def save_id(list):
    # transform into pandas DF
    property_id_df = pd.DataFrame(list)
    # drop duplicated observations
    property_id_df.drop_duplicates("id", inplace = True)
    # save data to excel
    property_id_df.to_excel('property_id_'+ date_string + '.xls')
    
save_id(property_id)
```

```python
property_characteristics_all = []
count = 0
for url in url_list:
    print(count/len(url_list)*100)
    property_data = BeautifulSoup(requests.get(url).content, "html.parser")
    property_tables = property_data.find_all("tr", class_="classified-table__row")
    property_characteristics = {"url":url}  
    for table in property_tables:
        text = [i.strip() for i in table.findAll(text=True)]
        key = text[1]
        value = text[3]
        property_characteristics[key] = value
    property_characteristics_all.append(property_characteristics)
    count += 1
```

```python
def save_characteristics(list):
    property_char_df = pd.DataFrame(list)
    property_char_df.to_excel('property_char_' + date_string +'.xls')

save_characteristics(property_characteristics_all)
```
